## reading and visualizing REE XML files

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import w2m.data
```

## DesgloseIntraSujeto

```python
f = "/Users/rlx/datasets/w2m/w2m_raw/REE/2018/05/desgloseintrawm01_2018050101.1.xml"
d = w2m.data.read_REE_DesgloseIntraSujeto(f)
d.head()
```

### valores encontrados en el Ã­ndice

### visualizaciÃ³n de algunos registros elegidos al azar

```python
idxs = np.random.permutation(range(len(d)))[:5]
for i in idxs:
    plt.plot(d.iloc[i].values, label=d.iloc[i].name)
    plt.xticks(range(len(d.iloc[i])), d.iloc[i].index, rotation="vertical");
plt.legend()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

### ActualizacionOfertasRestricciones

```python
filename = "/Users/rlx/datasets/w2m/w2m_raw/REE/2018/05/actualizacionofertasrestriccioneswm01_20180501.51.xml"
d = w2m.data.read_REE_ActualizacionOfertasRestricciones(filename)
```

```python
s = d.loc["EFWMARK"]
s
```

### visualizaciÃ³n de algunos registros elegidos al azar

```python
for _ in range(5):
    plt.figure(figsize=(20,3))
    for up in np.random.permutation(d.index)[:10]:
        s = d.loc[up]
        plt.subplot(141)
        plt.plot([i["baj"]["ctd"] for i in s])
        plt.xticks(range(len(s)), s.index, rotation="vertical");
        plt.title("cantidad bajada")
        plt.subplot(142)
        plt.plot([i["baj"]["precio"] for i in s])
        plt.xticks(range(len(s)), s.index, rotation="vertical");
        plt.title("precio bajada")
        plt.subplot(143)
        plt.plot([i["sub"]["ctd"] for i in s])
        plt.xticks(range(len(s)), s.index, rotation="vertical");
        plt.title("cantidad subida")
        plt.subplot(144)
        plt.plot([i["sub"]["precio"] for i in s], label=s.name)
        plt.xticks(range(len(s)), s.index, rotation="vertical");
        plt.title("precio subida")
    plt.legend()
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
```
