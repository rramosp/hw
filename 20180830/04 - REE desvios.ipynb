```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d = pd.read_csv("/home/rlx/Downloads/LIQUIDACIONES/DATOS/ZIP W2M/C2_reprdesv_20180501_18XWMARK-1234-18",
               delimiter=";", skiprows=2)
d.columns=[ "fecha", "periodo", "c1", "c2", "c3", "c4", "c5", "unidad", "nada"]
d = d[[i for i in d.columns if i!="nada"]].dropna()
```

```python
d.unidad.value_counts()
```

## algunas unidades aleatorias

```python
units = [i for i in  np.unique(d.unidad) if "WMARK" not in i]
units_wmark = [i for i in  np.unique(d.unidad) if "WMARK" in i]
len(units), len(units_wmark)
```

## unidades *WMARK

```python
plt.figure(figsize=(10,3))
for unit in units_wmark:
    k = d[d.unidad==unit]
    plt.plot(k.periodo, k.c1, label=unit)
plt.xlabel("periodo")
plt.ylabel("columna c1")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

## otras unidades

```python
for i in range(5):
    plt.figure(figsize=(10,3))
    for unit in units[i*10:(i+1)*10]:
        k = d[d.unidad==unit]
        plt.plot(k.periodo, k.c1, label=unit)
    plt.xlabel("periodo")
    plt.ylabel("columna c1")
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
plt.figure(figsize=(10,3))
for col in ["c%d"%i for i in range(1,6)]:
    k = d[d.unidad=="EFWMARK"]
    plt.plot(k.periodo, k[col], label="columna "+col)
plt.xlabel("periodo")
plt.ylabel("valor")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("EFWMARK")
```
