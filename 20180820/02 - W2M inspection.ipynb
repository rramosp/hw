```python
import w2m.data, w2m.utils
```

```python
cfg = w2m.utils.cfg2dict("config.ini")
w2m_rawdir = cfg["data"]["w2m_raw"]
omie_rawdir = cfg["data"]["omie_raw"]
print "using w2m raw data dir at ", w2m_rawdir
print "using omie raw data dir at", omie_rawdir
```

- **PDBC**: Programa diario básico de casación
- **PIBC**: Programa intra diario básico de casación
- **PIBCI**: Programa intra diario básico de casación registrado de forma **incremental**
- **PHF**: Programa horario final
    

## actividad por unidad de W2M

showing **energía de casación**:
- in a single specific day
- for all W2M units present that day
- **intraday** is **incremental** (+/-)

```python
pdbc_fmt = w2m_rawdir+"/OMIE/PROGRAMAS/pdbc_WMARK_%s.1"
pibc_fmt = w2m_rawdir+"/OMIE/PROGRAMAS/pibci_WMARK_%s%02d.1"

year, mes, dia = 2018, 5,1
```

```python
pdbc = w2m.data.read_pdbc(pdbc_fmt%("%d%02d%02d"%(year,mes,dia)))
for i in pdbc.unidad.unique():
    fig = plt.figure(figsize=(10,3))
    w2m.data.InfoUnidadDia.enery_from_casacion(pdbc_fmt, pibc_fmt, year, mes, dia,i).plot()    
```
